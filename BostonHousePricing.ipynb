```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
boston_data = pd.read_csv("data/boston_house_prices.csv", header=1)
boston_data
```

```python
boston_data.rename({'MEDV':'PRICE'}, axis=1, inplace=True)
boston_data
```

```python
boston_data.describe()
```

```python
boston_data.info()
```

```python
fig, ax = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
ax = ax.flatten()
for col, value in boston_data.items():
	sns.boxplot(y=col, data=boston_data, ax=ax[index])
	index +=1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
fig, ax = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
ax = ax.flatten()
for col, value in boston_data.items():
	sns.histplot(value, ax=ax[index])
	index +=1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```
