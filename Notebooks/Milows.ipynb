```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
boston_data = pd.read_csv("../data/boston_house_prices.csv", header=1)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(boston_data['MEDV'], bins=30, kde=True)
plt.title('Distribución de los Precios de las Viviendas')
plt.xlabel('Precio (en miles)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
correlation_matrix = boston_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='RM', y='MEDV', data=boston_data)
plt.title('Número de Habitaciones vs Precio')
plt.xlabel('Número de Habitaciones')
plt.ylabel('Precio (en miles)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='LSTAT', y ='MEDV', data=boston_data)
plt.title(' Estatus Inferior vs Precio')
plt.xlabel(' Estatus Inferior')
plt.ylabel('Precio (en miles)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MEDV', y ='CRIM', data=boston_data)
plt.title('precio vs crim')
plt.xlabel('Precio')
plt.ylabel('Criminalidad')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='MEDV', y ='LSTAT', data=boston_data)
plt.title('precio vs indice')
plt.xlabel('Precio')
plt.ylabel('indice')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=boston_data['MEDV'], color='sienna');
```
